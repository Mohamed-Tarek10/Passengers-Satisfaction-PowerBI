```python
pre_flight_cols = ['Ease of Online Booking', 'Check-in Service', 'Online Boarding', 'Gate Location', 'Baggage Handling']
data['Pre-flight Service Rate'] = df[pre_flight_cols].mean(axis=1)
```

```python

data["Pre-flight Service"] = np.where(
    data["Pre-flight Service Rate"] < 5, "Poor",
    np.where(
        data["Pre-flight Service Rate"] < 10, "Natural",
        np.where(
            data["Pre-flight Service Rate"] < 15, "Good",
            "Excellent"
        )
    )
)
data["Pre-flight Service"]
```

```python

data["Pre-flight Service Rate"] =  data["Ease of Online Booking"] + data["Check-in Service"] + data["Gate Location"] + data["Online Boarding"]
```

```python
# Creating the in-flight service rate column.
"""
 It will be based on the values in the columns ['On-board Service', 'Seat Comfort', 'Leg Room Service', 'Cleanliness',
       'Food and Drink', 'In-flight Service', 'In-flight Wifi Service','In-flight Entertainment']
"""
data["In-flight Service Rate"] = data["On-board Service"] + data["Seat Comfort"] + data["Leg Room Service"] + data["Cleanliness"] +data["Food and Drink"] + data["In-flight Service"] + data["In-flight Wifi Service"] + data["In-flight Entertainment"]
```

```python
# Creating a column for in-flight service.
data["In-flight Service"] = np.where(
    data["In-flight Service Rate"] < 10, "Poor",
    np.where(
        data["In-flight Service Rate"] < 20, "Natural",
        np.where(
            data["In-flight Service Rate"] < 30, "Good",
            "Excellent"
        )
    )
)
data["In-flight Service"]
```

```python
# Creaing the post-flight service rate column.
# It will based on the values in the columns ['Baggage Handling'], so, we can leave it the same.

# Then, we will compute the total service rate for the company.

data["Total Service"] = data["Pre-flight Service Rate"] + data["In-flight Service Rate"] + data["Baggage Handling"]
```

```python
# We can make a column for Age Group


data["Age Group"] = np.where(
    data["Age"] < 18, "Under 18",
    np.where(
        (data["Age"] >= 18) & (data["Age"] <= 25), "18-25",
        np.where(
            (data["Age"] >= 26) & (data["Age"] <= 35), "26-35",
            np.where(
                (data["Age"] >= 36) & (data["Age"] <= 45), "36-45",
            np.where(
                (data["Age"] >= 46) & (data["Age"] <= 55), "46-55",
                "+55"
            )

        )
    )
    )

)
data["Age Group"]
```

**Which services show the greatest rating differences between satisfied and dissatisfied passengers?**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


service_cols = [
    'Seat Comfort', 'Food and Drink', 'Gate Location',
    'In-flight Wifi Service', 'In-flight Entertainment', 'Online Boarding',
    'Leg Room Service', 'On-board Service', 'Check-in Service',
    'Cleanliness', 'Baggage Handling', 'In-flight Service'
]

service_analysis = df.groupby('Satisfaction')[service_cols].mean().T

service_analysis['Gap'] = service_analysis['Satisfied'] - service_analysis['Neutral or Dissatisfied']
service_analysis = service_analysis.sort_values(by='Gap', ascending=True)

plt.figure(figsize=(12, 8))

plt.hlines(y=service_analysis.index,
           xmin=service_analysis['Neutral or Dissatisfied'],
           xmax=service_analysis['Satisfied'],
           color='grey', alpha=0.4)


plt.scatter(service_analysis['Neutral or Dissatisfied'], service_analysis.index,
            color='orangered', alpha=1, s=100, label='Neutral or Dissatisfied')


plt.scatter(service_analysis['Satisfied'], service_analysis.index,
            color='teal', alpha=1, s=100, label='Satisfied')

plt.title(f"Page5: Service Gap Analysis (What Matters Most?)", fontsize=14)
plt.xlabel('Average Rating (1-5)', fontsize=12)
plt.ylabel('Airline Services', fontsize=12)
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

**How does trip distance impact satisfaction for Economy passengers compared to Business passengers?**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd



def classify_distance(dist):
    if dist < 1000: return 'Short Haul (<1k)'
    elif dist < 3000: return 'Medium Haul (1k-3k)'
    else: return 'Long Haul (>3k)'

df['Distance Category'] = df['Flight Distance'].apply(classify_distance)

dist_class_satisfaction = df.groupby(['Distance Category', 'Class'])['Satisfaction'].apply(lambda x: (x == 'Satisfied').mean() * 100).unstack()

dist_class_satisfaction = dist_class_satisfaction.reindex(['Short Haul (<1k)', 'Medium Haul (1k-3k)', 'Long Haul (>3k)'])

dist_class_satisfaction = dist_class_satisfaction[['Economy', 'Economy Plus', 'Business']]

plt.figure(figsize=(12, 7))
ax = dist_class_satisfaction.plot(kind='bar', figsize=(12, 7), colormap='viridis', edgecolor='black')

plt.title(f"Page 7: The Endurance Test\n(Does flight length kill satisfaction in Economy?)", fontsize=15, fontweight='bold')
plt.ylabel('Satisfaction Rate (%)', fontsize=12)
plt.xlabel('Flight Distance Category', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Travel Class')
plt.xticks(rotation=0)


for c in ax.containers:
    ax.bar_label(c, fmt='%.1f%%', label_type='edge', padding=3, weight='bold')

plt.tight_layout()
plt.show()
```

**Which nationalities represent the highest concentration of dissatisfied passengers?**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd




nat_satisfaction = pd.crosstab(df['Nationality'], df['Satisfaction'])

nat_satisfaction = nat_satisfaction[nat_satisfaction.sum(axis=1) > 100]

nat_satisfaction['Dissatisfied_Rate'] = (nat_satisfaction['Neutral or Dissatisfied'] / nat_satisfaction.sum(axis=1)) * 100


top_dissatisfied_nations = nat_satisfaction.sort_values('Dissatisfied_Rate', ascending=False).head(10)

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top_dissatisfied_nations['Dissatisfied_Rate'], y=top_dissatisfied_nations.index, palette='magma')


plt.title(f"Page 8: Geography of Complaints \n(Top 10 Nationalities with Highest Dissatisfaction Rates)", fontsize=16, fontweight='bold', color='#c0392b')
plt.xlabel('Dissatisfaction Rate (%)', fontsize=12)
plt.ylabel('Nationality', fontsize=12)

for i in ax.containers:
    ax.bar_label(i, fmt='%.1f%%', padding=3, fontweight='bold')

plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```

**Based on the Airport Name column, which airports generate the largest number of dissatisfied passengers?**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd




airport_counts = df['Airport Name'].value_counts()


valid_airports = airport_counts[airport_counts > 20].index

filtered_df = df[df['Airport Name'].isin(valid_airports)]
airport_satisfaction = pd.crosstab(filtered_df['Airport Name'], filtered_df['Satisfaction'])


airport_satisfaction['Dissatisfied_Rate'] = (airport_satisfaction['Neutral or Dissatisfied'] / airport_satisfaction.sum(axis=1)) * 100


worst_airports = airport_satisfaction.sort_values('Dissatisfied_Rate', ascending=False).head(20)




if not worst_airports.empty:
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x=worst_airports['Dissatisfied_Rate'], y=worst_airports.index, palette='Reds_r')

    plt.title(f"Page 9: The 'No-Fly' List \n(Airports with Highest Dissatisfaction Rates)", fontsize=15, fontweight='bold', color='darkred')
    plt.xlabel('Dissatisfaction Rate (%)')
    plt.ylabel('Airport Name')
    plt.xlim(0, 100)

    for i in ax.containers:
        ax.bar_label(i, fmt='%.1f%%', padding=3, fontweight='bold')

    plt.tight_layout()
    plt.show()

```

**Which factor has a stronger impact on satisfaction: improving Online Boarding or enhancing Gate Location?**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def get_satisfaction_rate(feature):
    ct = pd.crosstab(df[feature], df['Satisfaction'], normalize='index')
    if 'Satisfied' in ct.columns:
        return ct['Satisfied'] * 100
    else:
        return pd.Series(0, index=ct.index)

online_boarding_rates = get_satisfaction_rate('Online Boarding')
gate_location_rates = get_satisfaction_rate('Gate Location')


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.lineplot(x=online_boarding_rates.index, y=online_boarding_rates.values, marker='o', linewidth=3, color='green')
plt.title('Online Boarding (High Correlation)\nImprove this -> Satisfaction Skyrockets! ', fontsize=12, color='green', fontweight='bold')
plt.xlabel('Rating (1-5)')
plt.ylabel('% Satisfied')
plt.ylim(0, 100)
plt.grid(True)

plt.subplot(1, 2, 2)
sns.lineplot(x=gate_location_rates.index, y=gate_location_rates.values, marker='o', linewidth=3, color='grey')
plt.title('Gate Location (Low Correlation)\nDoes not matter much ', fontsize=12, color='grey', fontweight='bold')
plt.xlabel('Rating (1-5)')
plt.ylabel('% Satisfied')
plt.ylim(0, 100)
plt.grid(True)

plt.suptitle('Correlation Explained: Where to focus your money? ', fontsize=16)
plt.tight_layout()
plt.show()
```

**Does a longer flight make passengers less happy**

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt



df['Satisfaction_binary'] = df['Satisfaction'].apply(lambda x: 1 if x == 'Satisfied' else 0)

df['Distance_bin'] = pd.qcut(df['Flight Distance'], 10)

dist_sat = df.groupby('Distance_bin')['Satisfaction_binary'].mean()

plt.figure(figsize=(10, 5))
dist_sat.plot(kind='line', marker='o', color='purple', linewidth=2)

plt.title("Satisfaction Trend by Flight Distance 📏\n(Does Satisfaction drop as we fly further?)", fontsize=14, weight='bold')
plt.ylabel("Satisfaction Rate")
plt.xlabel("Flight Distance Ranges")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
